# This notebook will serve as a way to clean the remaining txt that will be used for char generation

```python
import re
```

```python
with open("corpus_RNN_Rohff.txt", "r", encoding="utf-8", errors="replace") as f:
    corpus = f.read()
```

As the task will be char generation, first I will let indications such as INTRO, etc... in order to maybe have a RNN that can reproduce the structure of a song

```python
corpus[:200]
```

```python
print(len(set(corpus)),sorted(set(corpus)))
```

### Characters : <>

```python
set(re.findall(r"\<.*\>", corpus)) #No need to modify this
```

### Characters : {}

```python
re.findall(r"\{.*?\}", corpus, flags=re.DOTALL)
```

These parts doesn't seems good so I will delete them entirely

```python
corpus = re.sub(r"\{.*?\}", "", corpus, flags=re.DOTALL)
```

### Characters : ()

```python
re.findall(r"\(.*?\)", corpus, flags=re.DOTALL)
```

```python
corpus = re.sub(r"\(.*?\)", "", corpus, flags=re.DOTALL)
re.findall(r"\(.*?\)", corpus, flags=re.DOTALL)
```

```python
print(len(set(corpus)),sorted(set(corpus)))
```

```python
re.findall(r"\w+\)", corpus, flags=re.DOTALL)
```

```python
corpus = re.sub(r"\)", "", corpus, flags=re.DOTALL)
```

### Character : *

```python
len(re.findall(r"\*\w+|\w+\*", corpus)), re.findall(r"\*\w+|\w+\*", corpus)[:10]
```

```python
def remove_star(match):
    return match.group(0).replace("*", "")

corpus = re.sub(r"\*\w+|\w+\*", remove_star, corpus, flags=re.DOTALL)
```

```python
len(re.findall(r"\*\w+|\w+\*", corpus, flags=re.DOTALL)), re.findall(r"\*\w+|\w+\*", corpus, flags=re.DOTALL)[:10]
```

Strange that this doesn't clean all the corresponding pattern

```python
corpus = re.sub(r"\*\w+", remove_star, corpus, flags=re.DOTALL)
corpus = re.sub(r"\w+\*", remove_star, corpus, flags=re.DOTALL)
re.findall(r"\*\w+|\w+\*", corpus, flags=re.DOTALL)
```

### Characters : « »

```python
re.findall("«.*?»", corpus)
```

```python
corpus = re.sub("«|»", "", corpus)
```

### Characters : “ ”

```python
re.findall(".*”", corpus)
```

```python
corpus = re.sub("”", "", corpus)
corpus = re.sub("“", "", corpus)
```

### Character : …

```python
re.findall(".*….*", corpus)
```

```python
corpus = re.sub("…", "...", corpus)
```

```python
print(len(set(corpus)),sorted(set(corpus)))
```

### Characters : ‘ ’

```python
re.findall(".*‘.*",corpus)
```

```python
corpus = corpus.replace("‘’", "")
```

```python
re.findall(".*’.*",corpus)
```

```python
def remove_element(match):
    return match.group(0).replace("‘", "'").replace("’", "'")
```

```python
corpus = re.sub("\w+‘\w+|\w+’\w+", remove_element, corpus)
```

```python
re.findall("\w+‘\w+|\w+’\w+",corpus)
```

```python
corpus = re.sub("\w+’\w+", "'", corpus)
re.findall("\w+‘\w+|\w+’\w+",corpus)
```

```python
re.findall(".*‘.*",corpus)
```

```python
corpus = re.sub("‘", "", corpus)
re.findall(".*‘.*",corpus)
```

```python
re.findall(".*’.*",corpus)
```

```python
corpus = re.sub("’", "", corpus)
re.findall(".*’.*",corpus)
```

### Characters : е (different than normal e)

```python
re.findall(".*е.*", corpus)[:5]
```

```python
corpus = re.sub("е", "e", corpus)
re.findall(".*е.*", corpus)[:5]
```

```python
print(len(set(corpus)),sorted(set(corpus)))
```

### Character : \x81

```python
re.findall(".*\x81.*", corpus)
```

```python
corpus = re.sub("\x81", "", corpus)
```

## Character : \x9c

```python
re.findall(".*\x9c.*", corpus)
```

```python
corpus = re.sub("\x9c", "", corpus)
```

Ok the txt seems good for now

```python
with open(f"clean_corpus_Rohff.txt", "w", encoding="utf-8") as f :
    f.write(corpus)
```
