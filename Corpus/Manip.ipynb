```python
import pickle
import numpy as np
import os
```

# Regroup the corpora

```python
#script_dir = os.path.dirname(__file__)
#path = os.path.join(script_dir, "..", "Corpus")
list_files = os.listdir("Cleaning/")

new_corp = []
for i in list_files :
    if "clean_corpus" in i  :
        with open(f"Cleaning\{i}", "r", encoding="utf-8", errors="replace") as f:
            corpus = f.read()
            new_corp.extend(corpus)
corpora = "".join(new_corp)
```

# Create encoding

```python
characs = sorted(set(corpora))
char2int = {ch: i for i, ch in enumerate(characs)}
```

```python
encode_corpora = [char2int[i] for i in corpora]
```

```python
np.save("Encoding_RNN_LSTM/corpora_encoded.npy",encode_corpora)
with open("Encoding_RNN_LSTM/encoding_map.pkl", "wb") as f:
    pickle.dump(char2int, f)
```
